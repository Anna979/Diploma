# Очистка данных


Добавить анализ выбросов

```python
data2018.shape
```

```python
def preprocessing(data):
    #replacing to nan values
    data.replace(['No Practice', 'No corporate income tax', '--'], np.nan, inplace=True)
    
    #deleting nan values
    data.dropna(thresh=data2018.shape[0]*0.8, axis=1, inplace=True)
    data.dropna(thresh=data2018.shape[1]*0.8, inplace=True)
    
    #filling in left nan values
    for column in data.iloc[:,3:].columns:
        data[column].fillna(data[column].median(), inplace=True)
```

```python
preprocessing(data2018)
```
